```python
from stan_util import get_stan_model,print_code
import arviz as az
N=4

model=get_stan_model("neal0")
print_code(model)
fit=model.sampling(data={"N":N},warmup=1000,iter=1500,chains=8)
print(fit)
az.plot_pair(fit,var_names=["x","y"])
```

```python
model=get_stan_model("neal1")
print_code(model)
fit=model.sampling(data={"N":N},warmup=1000,iter=1500,chains=8)
print(fit)
az.plot_pair(fit,var_names=["x","y"])
az.plot_pair(fit,var_names=["x_raw","y_raw"])
```
